```python
#抓取评论数的函数
import re
import json

commentURL='https://comment.sina.com.cn/page/info?version=1&format=json&channel=gn&\
newsid=comos-{}&group=undefined&compress=0&\
ie=utf-8&oe=utf-8&page=1&page_size=3&t_size=3&h_size=3'

def getCommentCount(newsurl):
    m=re.search('doc-i(.+).shtml',newsurl)
    newsid=m.group(1)
    comments=requests.get(commentURL.format(newsid))
    jd=json.loads(comments.text)
    return jd['result']['count']['show']

getCommentCount('https://news.sina.com.cn/c/2019-01-30/doc-ihqfskcp1826116.shtml')
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime

def getNewsDetail(newsurl):
    result={}
    res=requests.get(newsurl)
    res.encoding='utf-8'
    soup=BeautifulSoup(res.text)
    result['title']=soup.select('.main-title')[0].text
    result['newssource']=soup.select('.date-source a')[0].text
    time=soup.select('.date-source')[0].contents[1].text
    result['dt']=datetime.strptime(time,'%Y年%m月%d日 %H:%M')
    result['article']=' '.join([p.text.strip( ) for p in soup.select('#article p')[0:29]])
    result['editor']=soup.select('p.show_author')[0].text.strip('责任编辑：')
    result['comments']=getCommentCount(newsurl)
    return result

getNewsDetail('https://news.sina.com.cn/c/2019-01-30/doc-ihqfskcp1826116.shtml')
```
